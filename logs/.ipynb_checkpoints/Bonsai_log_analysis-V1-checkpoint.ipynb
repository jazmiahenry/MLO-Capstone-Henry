```python
filename = '2021-07-19-22-47-11_test_policy_log.csv'
df = pd.read_csv(filename)
df.head(5)
```

```python
# list of columns
df.columns
```

```python
# convert string to list
def str_to_lst(in_lst):
    res = in_lst.strip('][').split(', ')
    lst = list(np.array(res, dtype=int))
    return lst  
df_new = pd.DataFrame()
selected_columns = ['state_machines_speed', 'state_machines_state']
for col in selected_columns:
    df_new[col] = df[col].apply(str_to_lst)
```

```python
# flatten the columns
def flatten_column(df, col_lst):
    df_flat = pd.concat([pd.DataFrame(df[x].values.tolist()).add_prefix(x) for x in col_lst], axis=1)
    df = pd.concat([df_flat, df.drop(col_lst, axis=1)], axis=1)
    return df
```

```python
# Applying the flattening function
selected_columns = ['state_machines_speed', 'state_machines_state']
df_flattened = flatten_column(df_new, selected_columns)
```
